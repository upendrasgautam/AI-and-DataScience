# Descriptive Statistics
Descriptive statistics describe, show, and summarize the basic features of a dataset found in a given study, presented in a summary that describes the data sample and its measurements. It helps analysts to understand the data better.



## 1) Measure of Central Tendency
Central Tendency is used to indicate where does the middle or center of the distribution of our data lies. Central Tendency tells us where does the maturity of data lies.



## 2) Measures of Spread
Measure of Spread describes how wide the set of data is around the measure of central tendency/central value. 

Types of measure of Spread:

1) Range<br/>
2) Standard Deviaion<br/>
3) Variance<br/>
4) Quartiles<br/>

### 1) Range
Range tells us about the range of values in the set of data.
### 2) Standard Deviation
Standard Deviation is the measure of how much the data is spread out around the mean.
### 3) Variation
Variance is the whole square of Standard Deviation.
### 4) Quartiles
- It divides the given data set into quarters.
- Divides data in four segments based on the falling of numbers on the number line.

## Measure Of Symmetry
Measure of Symmetry is used to determine the symmetrical distribution of data set. A data set is said to be symmetric if it appears to be identical to the left and right side of the center point.
